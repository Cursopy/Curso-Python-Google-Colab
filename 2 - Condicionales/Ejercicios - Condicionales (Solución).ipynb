# Ejercicio 1.

Escribe un programa en Python usando condicionales que sea capaz de llevar a cabo el juego de **Piedra, Papel o Tijeras** entre un jugador real y la computadora.

1. Crea una variable llamada `opciones` la cual serÃ¡ de tipo `lista` y allÃ­ almacenaras las palabras **Piedra, Papel, Tijeras.**

2. Del mÃ³dulo `random` importa la funciÃ³n `randint()`.

3. A una variable llamada `computadora` asÃ­gnale un valor aleatorio de la variable `opciones`, es decir, `computadora = opciones[????]` dentro de los corchetes deberÃ¡s hacer uso de la funciÃ³n `randint()` para tomar aleatoriamente un elemento entre los Ã­ndices de la lista `opciones`.

4. Realiza el programa considerando lo siguiente:

  4.1. DeberÃ¡s solicitarle al usuario su opciÃ³n mediante una funciÃ³n `input()`.

  4.2. Si la opciÃ³n de la computadora es igual a la opciÃ³n del usuario, la salida del programa deberÃ¡ ser `Empate`.  

  4.3. Si la opciÃ³n de la computadora es distinta a la del usuario deberÃ¡s chequear quien ha ganado. Posteriormente, deberÃ¡s imprimir un mensaje que muestre si el usuario ganÃ³ o perdiÃ³. En caso de que la computadora haya ganado, mostrar su alternativa.

  4.4. Si el usuario ingresa una opciÃ³n incorrecta, deberÃ¡ mostrarle un mensaje que diga que la opciÃ³n es incorrecta y que debe intentar de nuevo.

*Pista: Recuerda que a la piedra solo le gana el papel, es decir que si el usuario escoge piedra, se deberÃ¡ chequear si la opciÃ³n de la computadora es papel. Si la alternativa de la computadora fuera papel entonces gana la computadora, de lo contrario gana el usuario. AnÃ¡logamente con el resto de las opciones.*

```
from random import randint
opciones = ['piedra', 'papel', 'tijeras']
computadora = opciones[randint(0, 2)]

jugador = input('Ingresa tu opciÃ³n: [Piedra, Papel, Tijeras]\n--> ').lower()

if jugador == computadora:
  print(f'\nEmpate! Ambos escogieron {jugador}.')

elif jugador == 'piedra':
  if computadora == 'papel':
    print(f'\nPerdiste! ðŸ˜ª la computadora ha sacado {computadora}ðŸ§»')
  else:
    print(f'\nGanaste! ðŸ¤— Aplastaste a la computadora.')

elif jugador == 'papel':
  if computadora == 'tijeras':
    print(f'\nPerdiste! ðŸ˜ª la computadora ha sacado {computadora}âœ‚')
  else:
    print(f'\nGanaste! ðŸ¤— Cubriste a la computadora.')

elif jugador == 'tijeras':
  if computadora == 'piedra':
    print(f'\nPerdiste! ðŸ˜ª la computadora ha sacado {computadora}.')
  else:
    print(f'\nGanaste! ðŸ¤— Cortaste a la computadora.')

else:
  print('\nOpciÃ³n incorrecta, intenta de nuevo.')
```

# Ejercicio 2.

Escribe un programa en Python para simular un cajero automÃ¡tico. Al momento de correr el programa deberÃ¡ mostrar un menÃº con 4 opciones:

1. Saldo actual
2. Retiro
3. Depositar
4. Salir

*   Cree una variable antes de toda la lÃ³gica del programa que se llame saldo y sea igual a 5000.
*   Las unidades para el dinero deberÃ¡n estar en dÃ³lares ðŸ’².
*   Desde la opciÃ³n 1 a la opciÃ³n 3 deberÃ¡ mostrar el saldo actual.
*   En la opciÃ³n de retiro deberÃ¡ ser capaz de corroborar que la cantidad no supera al saldo actual y tampoco es una cantidad negativa. AdemÃ¡s, deberÃ¡ mostrar la cantidad retirada y el saldo actual luego del retiro.
*   En la opciÃ³n de depÃ³sito deberÃ¡ corroborar que la cantidad no sea negativa. AsÃ­ mismo, debe mostrar el saldo depositado y el saldo total luego del depÃ³sito.
*   Asegurarse de que si el usuario ingresa una cantidad incorrecta se le muestre un mensaje acorde a la situaciÃ³n.

```
print('Cajero automÃ¡tico')
print('-----------------')
print('''
1. Saldo actual
2. Retiro
3. Depositar
4. Salir
''')

saldo = 5000
opcion = int(input('Ingresa tu opciÃ³n:\n--> '))

if opcion == 1:
  print(f'\nTu saldo actual es de: ${saldo}')

elif opcion == 2:
  print(f'\nTu saldo actual es de: ${saldo}')
  retiro = float(input('\nIngresa el monto a retirar:\n--> $'))

  if retiro > 0 and retiro <= saldo:
    print(f'\nHas retirado: ${retiro}')
    print(f'Saldo actual: ${saldo - retiro}')
  else:
    print('\nCantidad invÃ¡lida o superior al saldo disponible.')

elif opcion == 3:
  print(f'\nTu saldo actual es de: ${saldo}')
  deposito = float(input('\nIngresa la cantidad a depositar:\n--> $'))

  if deposito > 0:
    print(f'\nHas depositado: ${deposito}')
    print(f'Saldo actual: ${saldo + deposito}')
  else:
    print('\nCantidad invÃ¡lida')

elif opcion == 4:
  print('\nGracias por usar nuestro servicio. Hasta pronto!')

else:
  print('\nIngresa una opciÃ³n vÃ¡lida')
```
