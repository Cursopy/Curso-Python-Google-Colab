```
class Garaje():

  def __init__(self):
    self.carros = []

  def __str__(self):
    return f'El garaje tiene {len(self)} carros.'

  def __len__(self):
    return len(self.carros)

  def __getitem__(self, i):
    return self.carros[i]
```

```
ford = Garaje()
```

```
ford.carros
```
