````
class Perro():
  '''Plantilla para crear a un Perro'''

  def __init__(self, nombre_perro, raza_perro, vacunas):
    self.nombre = nombre_perro
    self.raza = raza_perro
    self.vacunas = vacunas

  def ladrar(self, nombre_amo):
    return f'Woof! Te amo, amo {nombre_amo}'
````

````
class Estudiante():
  '''Plantilla para crear estudiantes, 
  con su nota y calcular el promedio.'''

  def __init__(self, nombre: str, notas: list):
    self.nombre = nombre
    self.notas = notas

  def promedio(self):
    return sum(self.notas) / len(self.notas)
````

````
estudiante_1 = Estudiante('John', [19, 18, 20])
````

````
print(estudiante_1.nombre, estudiante_1.notas)
````

````
estudiante_1.promedio()
````

````
class Circulo():
  '''Plantilla para crear un circulo,
  y calcular su area.'''

  PI = 3.1416 #variable de clase

  def __init__(self, radio: float):
    self.radio = radio #variables de instancia

  def area(self):
    return f'El area del circulo es: {round(pow(self.radio, 2) * Circulo.PI, 2)}'
````

````
micirculo = Circulo(30.1)
micirculo.radio
````

````
micirculo.area()
````

````
class Termometro:

  def __init__(self, temp, limite):
    self.temperatura = temp
    self.limite = limite
  
  def comprobar_temp(self, nueva_temp):
    self.temperatura = nueva_temp

    if self.temperatura >= self.limite:
      return f'La temperatura es de {self.temperatura} ºC. Estado: Alarma!'
    else:
      return f'La temperatura es de {self.temperatura} ºC. Estado: Normal!'
````

````
termometro = Termometro(20, 40)
````

````
termometro.temperatura
````

````
termometro.limite
````

````
termometro.comprobar_temp(41)
````
