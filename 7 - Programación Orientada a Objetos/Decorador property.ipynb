```
class Estudiante():

  def __init__(self, nombre, universidad):
    self.nombre = nombre
    self.universidad = universidad
    self.notas = []
  
  def promedio(self):
    return sum(self.notas) / len(self.notas)

class Preparador(Estudiante):

  def __init__(self, nombre, universidad, salario):
    super().__init__(nombre, universidad)
    self.salario = salario

  @property
  def salario_mensual(self):
    return self.salario * 160
```

```
estudiante1 = Preparador('John Doe', 'MIT', 5)
```

```
estudiante1.salario_mensual
```
