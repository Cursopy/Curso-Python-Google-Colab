```
class Empleado():

  num_del_empleado = 0
  porcent_aumento = 1.20

  def __init__(self, nombre, apellido, salario):
    self.nombre = nombre
    self.apellido = apellido
    self.correo = f'{nombre}.{apellido}@Miempresa.com'
    self.salario = salario
    Empleado.num_del_empleado += 1

  def nombre_completo(self):
    return f'{self.nombre} {self.apellido}'

  def aplicar_aumento(self):
    self.salario *= Empleado.porcent_aumento

  @classmethod
  def pasar_datos(cls, empleado):
    nombre, apellido, salario = empleado.split('-')
    return cls(nombre, apellido, int(salario))
```

```
empleado1 = Empleado('Rebeca', 'DiGiacomo', 60000)
```

```
empleado1.nombre_completo()
```

```
empleado1.correo
```

```
empleado1.num_del_empleado
```
