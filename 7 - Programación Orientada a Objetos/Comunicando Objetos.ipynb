```
class Persona():

  def __init__(self, nombre, apellido, edad, pais):
    self.nombre = nombre
    self.apellido = apellido
    self.edad = edad
    self.pais = pais

  def __str__(self):
    return f'{self.nombre} {self.apellido}'
  
  def saludar(self):
    return f'Hola, mi nombre es {self}, mucho gusto!'
  
  def responder(self, objeto):
    return f'El gusto es mio {objeto}. Yo me llamo {self}.'
  
  def preguntar_edad_pais(self, objeto):
    return f'Lindo nombre {objeto.nombre}. Ahora, que edad tienes y de donde eres?'
  
  def responder_edad_pais(self, objeto):
    return f'Gracias! {objeto.nombre}. Tengo {self.edad} años y soy de {self.pais}.'
```

```
lorianna = Persona('Lorianna', 'Taglieri', 26, 'Italia')
```

```
print(lorianna)
```

```
arthuro = Persona('Arthuro', 'Martinez', 27, 'Venezuela')
```

```
print(arthuro)
```

```
lorianna.saludar()
```

```
arthuro.saludar()
```

```
lorianna.responder(arthuro)
```

```
arthuro.responder(lorianna)
```

```
lorianna.preguntar_edad_pais(arthuro)
```

```
arthuro.responder_edad_pais(lorianna)
```
