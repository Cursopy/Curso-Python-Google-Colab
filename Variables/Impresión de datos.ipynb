```
#.format()

# Determine la concentración de una solución de 20 ml de HCl sabiendo que
# la cantidad de moles es 0.000844 mol HCl.

moles_HCl = 0.000844                # mol
vol_HCl = 20 / 1000                 # mililitros(ml) --> litros(L)
con_HCl = moles_HCl / vol_HCl       # molar (M)

print(con_HCl)
```

```
print('Para una cantidad de moles igual a: {} mol de HCl y un volumen de: {} L. Su concentración es de: {} M'.format(moles_HCl, vol_HCl, con_HCl))
```

```
print(
    '''Para una cantidad de moles igual a: {a: ^30.3f} mol de HCl 

  y un volumen de: {b: ^10.3f} L. 

  Su concentración es de: {c: ^10.3f} M'''

    .format(a = moles_HCl, b = vol_HCl, c = con_HCl)
    )
```

```
# F-strings

print(
    f'''Para una cantidad de moles igual a: {round(moles_HCl, 3)} mol de HCl 

  y un volumen de: {vol_HCl: 5.3f} L. 

  Su concentración es de: {con_HCl: 5.3f} M'''

    )
```
