# Ejercicio 1

1. La Pascua se celebra el domingo inmediatamente después de la primera luna llena luego de la primavera. Debido a que su fecha incluye un componente lunar, la Pascua no tiene una fecha fija en el calendario gregoriano. En cambio, puede ocurrir en cualquier fecha entre el 22 de marzo y el 25 de abril. El mes y el día de Pascua se pueden calcular para un año determinado utilizando el algoritmo Anonymous Gregorian Computus, que se muestra a continuación:

  1. Establezca $a$ igual al resto cuando el año se divide por 19

  2. Establezca $b$ igual a la función floor cuando el año es dividido por 100
  3. Establezca $c$ igual al resto cuando el año se divide por 100
  4. Establezca $d$ igual a la función floor cuando $b$ es dividido por 4
  5. Establezca $e$ igual al resto cuando $b$ se divide por 4
  6. Establezca $f$ igual a la función floor de $\frac{b + 8}{25}$
  7. Establezca $g$ igual a la función floor de $\frac{b - f + 1}{3}$
  8. Establezca $h$ igual al resto cuando $19·a + b - d - g + 15$ se divide por 30
  9. Establezca $i$ igual a la función floor cuando $c$ es dividido por 4
  10. Establezca $k$ igual al resto cuando $c$ se divide por 4
  11. Establezca $l$ igual al resto cuando $32 + 2·e + 2·i - h - k$ se divide por 7
  12. Establezca $m$ igual a la función floor de $\frac{a + 11·h + 22·l}{451}$
  13. Establezca el mes igual a la función floor de $\frac{h + l - 7·m + 114}{31}$
  14. Establezca el día igual a uno más el resto cuando $h + l - 7·m + 114$ es dividido por 31

Escriba un programa que implemente el algoritmo Anonymous Gregorian Computus para calcular la fecha de Pascua. Su programa debe leer el año del usuario y luego mostrar un mensaje apropiado que incluya la fecha de Pascua en ese año.

```
from math import floor
año = int(input('Por favor ingrese el año: '))

# Paso 1.
a = año % 19

# Paso 2.
b = floor(año / 100)

# Paso 3.
c = año % 100

# Paso 4.
d = floor(b / 4)

# Paso 5.
e = b % 4

# Paso 6.
f = floor((b + 8) / 25)

# Paso 7.
g = floor((b - f + 1) / 3)

# Paso 8.
h = (19 * a + b - d - g + 15) % 30

# Paso 9.
i = floor(c / 4)

# Paso 10.
k = c % 4

# Paso 11.
l = (32 + 2 * e + 2 * i - h - k) % 7

# Paso 12.
m = floor((a + 11 * h + 22 * l) / 451)

# Paso 13.
mes = floor((h + l - 7 * m + 114) / 31)

# Paso 14.
dia = 1 + (h + l - 7 * m + 114) % 31

print(
    f''' Para el año {año} La Pascua ocurre en la siguiente fecha: {dia}/{mes}/{año} '''
)
```

2. Cree un programa que determine qué tan rápido viaja un objeto cuando se deja caer de cierta altura y toca el suelo. El usuario ingresará la altura desde la cual cae el objeto en metros (m). Debido a que el objeto se deja caer, su velocidad inicial es de $ 0 \frac{m}{s} $. Suponga que la aceleración debido a la gravedad es de $ 9.81 \frac{m}{s^{2}} $.

  Datos:
   * $V_{f}^{2} = V_{0}^{2} + 2·g·d$

```
# Ejercicio 2
from math import sqrt

GRAVEDAD = 9.81                                                                           # m/s

distancia = float(input('Por favor ingrese la distancia de lanzamiento (m):\n\n--> '))    # m

# Forma 1

velocidadFinal1 = (2 * GRAVEDAD * distancia) ** 0.5   # m/s

# Forma 2
velocidadFinal2 = sqrt(2 * GRAVEDAD * distancia)      # m/s

print(
    f'''\nCuando se deja caer un objeto desde una altura de {distancia:.1f} m. 
La velocidad final que alcanza es de {velocidadFinal1:.1f} m/s '''
)
```

# Ejercicio 3

3. Un polígono es regular si sus lados tienen la misma longitud y los ángulos entre todos los lados adyacentes son iguales. El área de un polígono regular se puede calcular usando la siguiente fórmula, donde $s$ es la longitud de un lado y $n$ es el número de lados:

  Datos:

  * $area = \frac{n·s^{2}}{4 · tan(\frac{π}{n})}$

  Escriba un programa que lea $s$ y $n$ del usuario y luego muestre el área de un polígono regular construido a partir de estos valores.

```
# Ejercicio 3

from math import tan, pi

s = float(input('Ingrese la longitud de un lado:\n--> '))
n = int(input('\nIngrese la cantidad de lados:\n--> '))

area = (n * s ** 2)/(4 * tan(pi/n))

print(
    f''' \n Un polígono regular de {n} lados y de longitud de lados igual a {s:.2f},
    poseé un área de {area:.2f} unidades '''
)
```
