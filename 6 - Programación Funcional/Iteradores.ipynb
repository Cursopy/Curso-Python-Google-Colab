# **Iteradores**

**Iterables:** Objetos por los cuales se puede iterar, se puede recorrer sus elementos 1 a 1 mediante un ciclo.

**Iterador:** Permiten recorrer los elementos de una secuencia pero accediendo a los valores 1 a 1 solo cuando se van necesitando. *LOS ITERADORES TAMBIÃ‰N SON ITERABLES.*

```
x = [0, 1, 2, 3, 4, 5]
y = iter(range(6))

print(type(x), type(y))
```

```
next(y)
```

```
next(y)
```

```
for elemento in x:
  print(elemento)
```

```
for elemento in y:
  print(elemento)
```

```
#Almacenan los mismos elementos pero en el caso de 'x', accedemos a los elementos
#mediante una lista, y en el caso de 'y' accedemos a ellos solo cuando los necesitamos.

#Los iteradores sirven para ahorrar memoria en los programas.
```

```
lista = [(6, 'a'), (1, 'j'), (4, 't'), (0, 'h'), (2, 'r')]
lista_inversa = reversed(lista)
print(lista_inversa)
```

```
# Zip()

alumnos = ['Daniela', 'Mariana', "Carla", 'Maria']
edades = [26, 27, 28, 29, 30]
notas = [10, 11, 12, 13, 14, 15, 16, 17]

zip(alumnos, edades, notas)
```

```
for nombre, edad, nota in zip(alumnos, edades, notas):
  print('El alumno {} tiene {} aÃ±os de edad y su calificacion es de {}'.format(nombre, edad, nota))
```

```
list(zip(alumnos, edades, notas))
```

```
dict(zip(alumnos, notas))
```

```
#map

def cuadrado(numero):
  return pow(numero, 2)

a = map(cuadrado, [1, 2, 3, 4, 5])
```

```
# Crear una lista que contenga los numeros del 1 al 10 y para aquellos
# numeros que sean pares devuelva la palabra 'par'.

def chequeo_par(numero):
  if numero % 2 == 0:
    return 'par'
  else:
    return numero

lista = map(chequeo_par, range(1, 11))
print(lista)
```

```
list(lista)
```

```
lista = list(map(lambda x: 'par' if x % 2 == 0 else x, range(1, 11)))
lista
```

```
list(map(lambda x: x[::-1], ['Brillo', 'Amarillo', 'Colores']))
```

```
#filter

tuple(filter(lambda x: True if len(x) % 2 == 0 else False, ['Brillo', 'Amarillo', 'Colores']))

#Devolvera aquellos elementos cuyo resultado de la funcion sean True. Es un 
#iterador lÃ³gico.
```
