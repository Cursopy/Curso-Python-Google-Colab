# Ejercicio

Escriba un programa en **Python** que sea capaz de generar contraseñas de gran seguridad de manera aleatoria.

Para ello siga los siguientes pasos:


1.   Del módulo `random` importe la función `sample(datos, k)`. La función `sample(datos, k)` toma dos parámetros, el primero serán los `datos o población` (debe ser un iterable), y el segundo parámetro `k` será la cantidad de elementos que deseamos tomar de los datos. La salida de la función `sample()` será una *lista* con los elementos seleccionados aleatoriamente de la muestra. Por ejemplo, si los `datos` consisten en 20 elementos y el parámetro `k` es igual a 4, la función `sample()` ***tomará 4 elementos aleatorios de los datos.***

[Documentación de la función sample() del módulo random](https://docs.python.org/3/library/random.html)

2.   Crea las siguientes variables:


*   `minusculas` --> esta variable deberá almacenar el alfabeto latino en minúsculas sin contar la letra ñ.
*   `mayusculas` --> esta variable deberá almacenar el alfabeto latino en mayúsculas sin contar la letra ñ.
*   `numeros` --> esta variable deberá almacenar una cadena de caracteres con los números del 0 al 9.
*   `simbolos` --> esta variable deberá almacenar una cadena de caracteres con diferentes símbolos, ejemplo: !"#$%&'()*+,-./:;<=>?@[\]^_´{|}~ 

3.   Cree una variable llamada `caracteres` la cual almacenará la concatenación de las cuatro variables creadas en el punto 2.

4.   Cree una variable llamada `longitud` que almacenará la longitud que deberá tener la contraseña, por ejemplo: 16.

5.   Cree una variable llamada `clave` que haga uso de la función `sample()` para generar una contraseña empleando la variable `caracteres` como `población` y la variable `longitud` como `k`.


```
import random

minusculas = 'abcdefghijklmnopqrstuvwxyz' # No aparece la ñ
mayusculas = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
numeros = '0123456789'
simbolos = '''!"#$%&'()*+,-./:;<=>?@[\]^_´{|}~"'''

caracteres = minusculas + mayusculas + simbolos + numeros
longitud = 16
clave = "".join(random.sample(caracteres, longitud))

print(f'La contraseña es: {clave}\n\nPosee una longitud \
de {len(clave)} caracteres.')
```

```
from string import digits, punctuation, ascii_letters

caracteres1 = ascii_letters + digits + punctuation
clave1 = "".join(random.sample(caracteres1, longitud))

print(f'La contraseña es: {clave1}\n\nPosee una longitud \
de {len(clave1)} caracteres.')
```
