```
# Cinco amigos harán una carrera:
# 'Valeria', 'Daniela', 'Andres', 'Paola', 'Manuel'.

# Realiza un código en Python que sea capaz de determinar todos los posibles
# resultados de la carrera, es decir, el orden en el que pueden llegar
# a la meta los cinco amigos.

5 * 4 * 3 * 2 * 1
```

```
amigos = ['Valeria', 'Daniela', 'Andres', 'Paola', 'Manuel']
contador = 0

for a1 in amigos:
  for a2 in amigos:
    for a3 in amigos:
      for a4 in amigos:
        for a5 in amigos:
          if a1 != a2 and a1 != a3 and a1 != a4 and a1 != a5 and \
          a2 != a3 and a2 != a4 and a2 != a5 and \
          a3 != a4 and a3 != a5 and a4 != a5:
            orden = [a1, a2, a3, a4, a5]
            contador += 1
            print(f'{contador:3}) {orden}')
```

```
import itertools
amigos = ['Valeria', 'Daniela', 'Andres', 'Paola', 'Manuel']
contador = 0
permutacion = itertools.permutations(amigos, 5)

for i in permutacion:
  contador += 1
  print(f'{contador}) {i}')
```
