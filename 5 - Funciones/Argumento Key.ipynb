```
#Ordenar la lista de menor a mayor teniendo en cuenta primero el tercer
#elemento, luego el segundo y finalmente el primero.

datos = [(1, 3, 6), (7, 5, 9), (4, 7, 1), (5, 8, 0), (9, 8, 3)]
```

```
def ordenar(tupla):
  return (tupla[2], tupla[1], tupla[0])
```

```
a = (1, 3, 6)
ordenar(a)
```

```
datos.sort(key = ordenar)
```
