```
from typing import List, Dict, Union

Reactor = List[Dict[str, Union[int, float]]]
def reactor(temp: int, presion: int, conversion: float) -> Reactor:
  return [{
      'Temperatura': temp,
      'Presion': presion,
      'Conversion': conversion
  }]
```

```
Reactor_continuo = reactor(573, 10, 0.1)
print(Reactor_continuo)
```

```
reactor(1.0, 1, 0.3)
```

```
reactor()
```
