```
def persona(nombre, edad):
  print(nombre, edad)
```

```
persona('ronny', 26)
```

```
def persona1(**kwargs):
  print(kwargs)
```

```
# Recibe un par [parametro (clave) - argumento (valor)] y lo empaqueta
# en un diccionario.
```

```
persona1(nombre = 'Ronny', edad = 26, ciudad = 'caracas')
```

```
diccionario = {
    'nombre': 'Genesis',
    'edad': 30
}
```

```
def persona2(nombre, edad):
  print(nombre, edad)
```

```
persona2(**diccionario)
```

```
# Las claves tienen que coincidir con los parametros (en cuanto a nombre).
# La cantidad de elementos del diccionario tiene que ser la misma
# cantidad que recibe la funci√≥n. 
```

```
def ambos(*args, **kwargs):
  print(args)
  print(kwargs)
```

```
ambos(1, 2, 3, 4, 5, 6, [1, 2, 3], 'hola', nombre = 'John', apellido = 'Doe')
```
