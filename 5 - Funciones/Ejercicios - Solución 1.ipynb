# Ejercicio 1

Un isógrama es una palabra que no tiene letras repetidas. Implementa una función que determine si una cadena de caracteres que contiene *solo letras* es o no un isógrama. Su función deberá devolver `True` o `False`. Asuma que una cadena de carácter vacía es un isógrama y además ignore las mayúsculas.

```
Funcion("Cambur" ) -> True
Funcion("Aba" ) -> False
Funcion("CaRAcaS" ) -> False
```

````
def isograma(palabra):
  return len(palabra) == len(set(palabra.lower()))

isograma('')
````

# Ejercicio 2

Jaden Smith, hijo de Will Smith, es el protagonista de películas como The Karate Kid (2010) y After Earth (2013). Jaden también es conocido por su filosofía que transmite a través de Twitter. Cuando él escribe en Twitter, casi siempre usa mayúsculas en cada palabra, por ejemplo:

`"How Can Mirrors Be Real If Our Eyes Aren't Real"`

Tu tarea será crear una función que sea capaz de convertir cualquier cadena de caracteres a cómo las escribiría Jaden Smith.

````
from string import capwords

def jadensmith(frase):
  lista = []
  for palabra in frase.split():
    lista.append(palabra.capitalize())
  return ' '.join(lista)

def jadensmith2(frase):
  return capwords(frase)

jadensmith2('How can mirrors be real if our eyes aren\'t real')
````

# Ejercicio 3

Se te proporciona una lista (que tendrá una longitud de al menos 3 unidades, pero podría ser muy grande) que contiene números enteros. La lista estará compuesta por únicamente `enteros impares` o enteramente compuesta por `enteros pares` **excepto por un único entero N**. Escriba una función que tome la lista como argumento y devuelva este "valor atípico" N.

`[2, 4, 0, 100, 4, 11, 2602, 36]`

Deberá devolver: 11 (el único número impar)

---

`[160, 3, 1719, 19, 11, 13, -21]`

Deberá devolver: 160 (el único número par)

````
def atipico(lista):
  atipico_par = [N for N in lista if N % 2 == 0]
  atipico_impar = [N for N in lista if N % 2 == 1]
  return atipico_par if len(atipico_par) < len(atipico_impar) else atipico_impar

atipico([160, 3, 1719, 19, 11, 13, -21])
````

````
lista = [160, 3, 1719, 19, 11, 13, -21]
print([N for N in lista if N % 2 == 0])
print([N for N in lista if N % 2 == 1])
````

# Ejercicio 4

Crea una función para que convierta una cadena palabras delimitadas por guiones y/o guiones bajos al estilo **camelCase**.

`Funcion("El-guerrero-de-Acero")` 

Devolverá: `"elGuerreroDeAcero"`

`Funcion("el_guerrero_dorado")`

Devolverá: `"elGuerreroDorado"`
 
 
***Considere también que una misma cadena de palabras puede estar delimitada tanto por guiones como guiones bajos.***

`"Los-Caballeros_del_Zodiaco"`

````
def camelCase(frase):
  frase = frase.replace('-', ' ').replace('_', ' ').split()
  if len(frase) == 0:
    return ''
  return frase[0].lower() + ''.join([x.capitalize() for x in frase[1:]])

camelCase("")
````

# Ejercicio 5

Escribe una función en Python que reciba dos listas, `lista1` y `lista2`. Ambas listas van a contener única y exclusivamente palabras. La cuestión es que la `lista2` va a contener palabras que se hayan formado utilizando a lo sumo una palabra de la `lista1`. Tu función deberá devolver la(s) palabra(s) de la `lista1` se usaron para formar la(s) palabra(s) de la `lista2`.

Ejemplo:


```
Entrada:

lista1 = ['logo', 'logo', 'mar', ella]
lista2 = ["maravilla", "dermatólogo", "carne", "altamar", "botella"]
```



```
Salida:
['logo', 'mar', 'ella']
```

````
lista1 = ['logo', 'logo', 'mar', 'ella']
lista2 = ["maravilla", "dermatólogo", "carne", "altamar", "botella"]

def arreglo(lista1, lista2):
  lista = []
  for palabra1 in lista1:
    for palabra2 in lista2:
      if palabra1 in palabra2 and not palabra1 in lista:
        lista.append(palabra1)
  
  return lista

arreglo(lista1, lista2)
````

````
def arreglo2(lista1, lista2):
  return list({palabra1 for palabra1 in lista1 if any([palabra1 in palabra for palabra in lista2])})

arreglo2(lista1, lista2)
````
