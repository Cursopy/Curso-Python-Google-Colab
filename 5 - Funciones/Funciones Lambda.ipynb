````
import math
def area_circulo(radio: float) -> float:
  return math.pi * radio ** 2
````

````
area1 = area_circulo(5)
print(area1)
````

````
area = lambda r: math.pi * r ** 2
````

````
type(area)
````

````
area(5)
````

````
(lambda r: math.pi * r ** 2)(7)
````

````
def division(x, y):
  return x // y

z = division(4, 2)
print(f'El valor de z es: {z}\n')
````

````
w = lambda x, y: x / y
print(f'El valor de w es: {w(15, 3)}\n')
````

`# lambda parámetros: expresión if condición else condición`

Cuando se usan funciones lambda y se emplea el condicional if, es obligatorio usar el bloque else.

````
#Ordenar la siguiente lista tomando en cuenta el primer elemento de las tuplas.
data = [(6, 'R'), ('B', 3), (8, 'Y'), (13, 'J'), ('F', 24), ('N', 29)]
````

````
data.sort(key = lambda elemento: elemento[0] if isinstance(elemento[0], int) else elemento[1])
data
````

````
#Ordenar la lista primero con los multiplos de 10 y luego con los no multiplos
#de 10. Asi mismo, los numeros deberan estar ordenados.

datos = [20, 40, 41, 11, 51, 61, 30, 50, 21, 10, 31]
````

````
datos.sort()
````

````
datos
````

````
datos.sort( key = lambda x: x if x % 10 == 0 else x*10)
datos
````

````
datos1 = datos.copy()
````

````
a = [x if x % 10 == 0 else x*10 for x in datos1]
a.sort()
a
````
