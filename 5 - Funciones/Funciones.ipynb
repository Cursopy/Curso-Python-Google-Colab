````
def operaciones_alg(numero1, numero2):

  ''' Esta funcion le permitira al usuario realizar 
  operaciones algebraicas como la suma, la resta, 
  la multiplicacion y la division'''

  operacion = input('''Qué operación deseas realizar?
  
                Presiona "+" para sumar.
                Presiona "-" para restar.
                Presiona "*" para multiplicar.
                Presiona "/" para dividir.
   ''')
  
  if operacion == "+":
    return numero1 + numero2
  if operacion == "-":
    return numero1 - numero2
  if operacion == "*":
    return numero1 * numero2
  if operacion == "/":
    return numero1 / numero2
````

````
#Determinar si un numero dado por el usuario es primo
# 17 -> 1, 17
# 3 -> 1, 3
# 5 -> 1, 5
import math
def primo(n):

  if n == 1:
    return 'No es primo'
  
  raiz = int(math.sqrt(n))
  
  for i in range(2, raiz + 1):
    if n % i == 0:
      return 'No es primo'
  
  return 'Es primo'
````

````
import time

inicio = time.time()
a = primo(59)
print(a)
final = time.time()

print(final - inicio)
````

````
def encontrar_primos(inicio, final):

  primos = []

  if inicio < 2:
    inicio = 2

  for n in range(inicio, final + 1):
    for i in range(2, n):
      if n % i == 0:
        break
    else:
      primos.append(n)
  
  return primos
````

````
primos = encontrar_primos(1, 1000)
print(primos)
````

````
# Realizar un programa en Python que solicite una temperatura,
# y le de la opción al usuario de convertir ese número
# en grados centrigrados, fahrenheit y kelvin.

def centi_fahr(c):
  return (c * 1.8) + 32

def centi_kelvin(c):
  return c + 273.15

def fahr_centi(f):
    return (f - 32) / 1.8

def fahr_kelvin(f):
    return (5/9) * (f - 32) / 1.8 + 273.15

def kelvin_centi(k):
    return k - 273.15

def kelvin_fahr(k):
    return 1.8 * (k - 273.15) + 32

def menu():
  print('Conversor de temperaturas')
  print('-------------------------')
  print('1. Centigrados a Fahrenheit')
  print('2. Centigrados a Kelvin')
  print('3. Fahrenheit a Centigrados')
  print('4. Fahrenheit a Kelvin')
  print('5. Kelvin a Centigrados')
  print('6. Kelvin a Fahrenheit')
  print('7. Salir')
  opcion = int(input('\nIngresa tu opcion:\n--> '))
  return opcion

def conversion(opcion, simbolos):
  temp = int(input(f'\nIngresa una temp. en ({simbolos[opcion][0]}):\n-->'))
  return simbolos[opcion][2](temp)

def programa():

  simbolos = {
      1: ['ºC', 'ºF', centi_fahr],
      2: ['ºC', 'K', centi_kelvin],
      3: ['ºF', 'ºC', fahr_centi],
      4: ['ºF', 'K', fahr_kelvin],
      5: ['K', 'ºC', kelvin_centi],
      6: ['K', 'ºF', kelvin_fahr]
  }

  while True:
    opcion = menu()

    if opcion == 7:
      print('\nHasta pronto!\n')
      break
    
    elif opcion >= 1 and opcion <= 6:
      solucion = conversion(opcion, simbolos)
      print(f'\nSon {solucion}{simbolos[opcion][1]}\n\n')
    
    else:
      print('\nIngresa una opcion valida!\n')

programa()
````
