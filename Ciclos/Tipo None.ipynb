```
# Hagamos un programa que determine el numero mayor entre 5 numeros
# Dados por el usuario

contador = 1
mayor = None

while contador <= 5:
  numero = int(input('Ingresa un numero:\n--> '))
  contador += 1

  if mayor is None or numero > mayor:
    mayor = numero
else:
  print(f'\nEl numero mayor es {mayor}')
```
