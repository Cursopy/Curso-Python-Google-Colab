```
temp_C = (23, 41, 56, 63, 23, 0)
# (0 °C × 9/5) + 32 = 32 °F

for temperatura in temp_C:
  temp_F = (temperatura * 9/5) + 32
  print(temp_F)
```

```
numero = int(input('Ingresa un numero:\n--> '))

for n in range(11):
  print(f'{n}) {numero} x {n} = {numero * n}')
```

```
#for-in > El operador in del ciclo for es para ingresar a los elementos del iterable.
#operador in > Operador lógico.

meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio'
          'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

for mes in meses:
  if 'e' in mes:
    print(mes)
```

```
Presiones = [1.23, 1.0, 7.2, 8.1] #atm

for presion in Presiones:
  if 7.2 == presion:
    break
  else:
    print(presion)
```

```
numeros = [3, -7, 23, -2, 6, 8, 0]

for j in numeros:
  if j > 0:
    print('El numero', j, 'es positivo.')
  elif j < 0:
    print('El numero {} es negativo.'.format(j))
  else:
    continue
```

```
# Escribe un programa que imprima los números del 1 al 30. 
# Aquellos números que sean múltiplos de 3 el programa deberá sustituir 
# el número por la palabra "Fizz". Para aquellos que sean multiplos de 5, 
# deberá imprimir la palabra "Buzz". 
# Para aquellos números que sean multiplos de 3 y de 5 que imprima "FizzBuzz".

for numero in range(1, 31):
  if numero % 3 == 0 and numero % 5 == 0:
    print('FizzBuzz')
  elif numero % 3 == 0:
    print('Fizz')
  elif numero % 5 == 0:
    print('Buzz')
  else:
    print(numero)
```

```
amigos = ['John', 'Lorianna', 'Arthuro', 'Engi']

# Usando la funcion enumerate()

for indice, valor in enumerate(amigos):
  print(f'Amigo #{indice} es {valor}')

print('')
# Usando la funcion range()

for i in range(len(amigos)):
  print(f'Amigo #{i} es {amigos[i]}')
```

```
list(enumerate(amigos))
```

```
lista = (12, 412, 345, 21, 565, 23, 22)

for indice, valor in enumerate(lista):
  print(f'el numero {valor} se encuentra en el indice {indice} de la lista.')
  print(f'el numero {valor} se encuentra en la posicion {indice + 1} de la lista.')
  print('')
```

```
diccionario = {
    'Nombre': 'Ronny',
    'Edad': 26,
    'Pais': 'Venezuela'
    }

for i in diccionario:
  print(i)
```
