```
# Tabla de multiplicar 7
# 0) 7 x 0 = 0
# 1) 7 x 1 = 7
# 2) 7 x 2 = 14
# 3) 7 x 3 = 21
# 4) 7 x 4 = 28
# 5) 7 x 5 = 35
# 6) 7 x 6 = 42
# 7) 7 x 7 = 49
# 8) 7 x 8 = 56
# 9) 7 x 9 = 63

TABLA = 7
contador = 0

while contador <= 9:
  print(f'{contador}) {TABLA} x {contador} = {TABLA * contador}')
  contador += 1
```

```
numero = 7
condicion = True

while condicion:
  numero_usuario = int(input('Ingresa un numero:\n--> '))

  if numero_usuario != numero:
    print('\nTe has equivocado')
    continue # Hacer otra iteracion
    print('Hola')
  else:
    print('\nHas adivinado')
    break
    #condicion = False

else:
  print('Eres un Genio')



# Continue, Break, Else
```

```

condicion_usuario = True

while condicion_usuario:
  usuario = input('Ingresa tu nombre de usuario:\n--> ')

  if usuario == 'John':
    print('\nUsuario correcto.')
    print('-----------------')
    condicion_usuario = False
    condicion_clave = True
    intentos_clave = 3

    while condicion_clave:
      clave = input('\nIngresa tu clave:\n--> ')

      if clave == 'ronny123':
        print('\nClave correcta.')
        print('-----------------')
        print('\nBienvenido al sistema.')
        condicion_clave = False
      else:
        print('\nClave incorrecta, intenta de nuevo.')
        intentos_clave -= 1
        print(f'\nTe restan {intentos_clave} intentos.')
      
      if intentos_clave == 0:
        print('\nClave bloqueada.')
        break
  else:
    print('\nUsuario incorrecto, intenta de nuevo.')
```
