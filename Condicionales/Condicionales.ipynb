````
n1 = float(input('Ingresa un numero: \n -->'))
n2 = float(input('\n Ingresa otro numero:\n-->'))

if n1 > n2:
  print(f'\n {n1} es mayor que {n2}')

elif n2 > n1:
  print(f'\n {n2} es mayor que {n1}')

else:
  print(f'\n Los numeros son iguales')
````

````
# Adivinar un numero

numero = 6

jugar = input('Presiona "a" si quieres jugar: \n --> ').lower()

if jugar =='a': # Condicional principal
  numero_usuario = int(input('\nIngresa un numero:\n --> '))

  if numero_usuario == numero: # Condicional secundario
    print('\nHas adivinado! :)')

  # Cuando el usuario se haya equivocado en adivinar el numero por 1
  # Se le brindara esa pista.
  elif abs(numero_usuario - numero) == 1:
    print('\n Te equivocaste por 1')

  else:
    print('\nLo siento, te equivocaste!')

else:
  print('\n Hasta la proxima')
````
